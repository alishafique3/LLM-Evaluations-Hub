## Example 1: Accuracy
Compute accuracy for a classification task.

```python
from datasets import load_dataset, load_metric

# Load a sample dataset
dataset = load_dataset('glue', 'mrpc', split='train[:10%]')

# Let's assume these are the predicted and actual labels
predictions = [0, 1, 0, 1, 1, 0, 0, 1, 0, 1]
references = dataset['label'][:10]

# Load the accuracy metric
accuracy_metric = load_metric('accuracy')

# Compute the accuracy
accuracy = accuracy_metric.compute(predictions=predictions, references=references)

# Print the result
print("Accuracy:", accuracy)
```

## Example 2: Precision, Recall, F1 Score
Compute precision, recall, and F1 score for a classification task.

```python
from datasets import load_dataset, load_metric

# Load a sample dataset
dataset = load_dataset('glue', 'mrpc', split='train[:10%]')

# Let's assume these are the predicted and actual labels
predictions = [0, 1, 0, 1, 1, 0, 0, 1, 0, 1]
references = dataset['label'][:10]

# Load the precision, recall, and f1 metrics
precision_metric = load_metric('precision')
recall_metric = load_metric('recall')
f1_metric = load_metric('f1')

# Compute the precision, recall, and f1 score
precision = precision_metric.compute(predictions=predictions, references=references, average='binary')
recall = recall_metric.compute(predictions=predictions, references=references, average='binary')
f1 = f1_metric.compute(predictions=predictions, references=references, average='binary')

# Print the results
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

## Example 3: BLEU Score
Compute the BLEU score for a machine translation task.

```python
from datasets import load_dataset, load_metric

# Load a sample dataset
dataset = load_dataset('wmt14', 'de-en', split='test[:1%]')

# Let's assume these are the predicted translations and references
predictions = predictions = [
    ["hello", "there", "general", "kenobi"],
    ["foo", "bar", "foobar"]
]
references = [
    [["hello", "there", "general", "kenobi"], ["hello", "there", "!"]],
    [["foo", "bar", "foobar"]]
]

# Load the BLEU metric
bleu_metric = load_metric('bleu')
print(bleu_metric.inputs_description)

# Compute the BLEU score
bleu = bleu_metric.compute(predictions=predictions, references=references)

# Print the result
print("BLEU Score:", bleu)
```

## Example 4: ROUGE Score
ROUGE score for a summarization task. To be able to use rouge, you need to install the following dependency: rouge_score.

```python
!pip install rouge_score --quiet
```

```python
from datasets import load_dataset, load_metric

# Let's assume these are the generated summaries and references
predictions = ["The quick brown fox jumps over the lazy dog"]
references = ["A fast, brown fox leaps over a sleepy dog"]

# Load the ROUGE metric
rouge_metric = load_metric('rouge')
print(rouge_metric.inputs_description)

# Compute the ROUGE score
rouge = rouge_metric.compute(predictions=predictions, references=references)

# Print the result
print("ROUGE Score:", rouge)
```
